# Analysis of Blood Pressure (Systolic Readings)

The program analyzes spread out of values in a data set around the mean (average) of that set. Low standard deviation suggests data points are consistently close to the mean, indicating reliability and predictability. High standard deviation suggests variability, indicating that the data points are spread out and less predictable.

Here blood pressure study of systolic readings of two individuals indicate that one is stable and other is heavily fluctuating causing greater risk of heart attack.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data for blood pressure readings
np.random.seed(0)  # For reproducibility

# Patient 1: Low Standard Deviation
mean_bp1 = 125  # Mean blood pressure
std_dev1 = 5    # Low standard deviation
bp_readings1 = np.random.normal(mean_bp1, std_dev1, 100)

# Patient 2: High Standard Deviation
mean_bp2 = 125  # Mean blood pressure
std_dev2 = 25   # High standard deviation
bp_readings2 = np.random.normal(mean_bp2, std_dev2, 100)



# Plotting the data
plt.figure(figsize=(12, 6))

# Low standard deviation plot
plt.subplot(1, 2, 1)
plt.hist(bp_readings1, bins=15, color='blue', alpha=0.7, label='Low Std Dev')
plt.axvline(mean_bp1, color='red', linestyle='dashed', linewidth=1, label='Mean (120)')
plt.title('Blood Pressure Readings with Low Standard Deviation')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.legend()

# High standard deviation plot
plt.subplot(1, 2, 2)
plt.hist(bp_readings2, bins=15, color='Red', alpha=0.7, label='High Std Dev')
plt.axvline(mean_bp2, color='red', linestyle='dashed', linewidth=1, label='Mean (120)')
plt.title('Blood Pressure Readings with High Standard Deviation')
plt.xlabel('Blood Pressure (mmHg)')
plt.ylabel('Frequency')
plt.legend()

plt.tight_layout()
plt.show()
```
